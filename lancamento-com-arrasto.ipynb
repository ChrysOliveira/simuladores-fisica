```python
import matplotlib.pyplot as plt
import numpy as np

massa_bola = 0.454          
gravidade = 9.81          
velocidade = 29.1          
dt = 0.001
angulo = np.radians(17.6)
tempo = [0]                        
x = [0]                        
y = [0]

# densidade_ar = 1.05
# seccao_transversal = np.pi * 11**2

def defineCA(velocidade):
  if velocidade > 0 and velocidade < 20:
    return 0.5
  elif velocidade > 20:
    return 0.1

def defineFA(velocidade):
  return defineCA(velocidade)*velocidade**2 

vx = [velocidade*np.cos(angulo)]  
vy = [velocidade*np.sin(angulo)]
ax = [-(defineFA(velocidade)*np.cos(angulo))/massa_bola ]          
ay = [-gravidade-(defineFA(velocidade)*np.sin(angulo)/massa_bola) ]


count = 0
while (y[count] >= 0):                  
    tempo.append(tempo[count]+dt)               
        
    vx.append(vx[count]+dt*ax[count])  
    vy.append(vy[count]+dt*ay[count])

    x.append(x[count]+dt*vx[count])    
    y.append(y[count]+dt*vy[count])    

    velocidade = np.sqrt(vx[count+1]**2 + vy[count+1]**2)

    ax.append(-(defineFA(velocidade)*np.cos(angulo))/massa_bola)     
    ay.append(-gravidade-(defineFA(velocidade)*np.sin(angulo)/massa_bola))
    
    count = count +1


plt.plot(x,y)
plt.xlabel("distancia (m)")
plt.ylabel("altura (m)")
   

```
